```python
def sum69(nums):
    total = 0
    skip = False
    
    for n in nums:
        if n == 6:
            skip = True   # start ignoring
        elif skip and n == 9:
            skip = False  # stop ignoring
        elif not skip:
            total += n
    
    return total

# Test cases
print(sum69([1, 2, 2]))            # 5
print(sum69([1, 2, 2, 6, 99, 9]))  # 5
print(sum69([1, 1, 6, 7, 9, 2]))   # 4
print(sum69([]))                   # 0
```

```python
def spy_game(nums):
    code = [0, 0, 7]   # sequence weâ€™re looking for
    idx = 0            # index to track progress in code
    
    for n in nums:
        if n == code[idx]:
            idx += 1
            if idx == len(code):   # found all elements in order
                return True
    return False

# Test cases
print(spy_game([1,2,4,0,0,7,5]))     # True
print(spy_game([1,0,2,4,0,5,7]))     # True
print(spy_game([1,7,2,0,4,5,0]))     # False
```

```python
def spy_game(nums):
    code = [0,0,7,'x']

    for num in nums:
        if num == code[0]:
            code.pop(0)

    return len(code) == 1


print(spy_game([1,2,4,0,0,7,5,0]))
print(spy_game([1,0,2,4,0,5,7]))
print(spy_game([1,7,2,0,4,5,0]))
```

```python
def count_primes(n):
    if n < 2:
        return 0
    
    primes = []
    for num in range(2, n+1):
        is_prime = True
        for i in range(2, int(num**0.5) + 1):
            if num % i == 0:
                is_prime = False
                break
        if is_prime:
            primes.append(num)
    
    print(primes)
    return len(primes)
   

# Example usage
print(count_primes(100))  # Output: 25
```
